```python
import torchvision.transforms as T
from PIL import Image
from pathlib import Path
import pandas as pd
import numpy as np


tranform = T.ColorJitter(
    brightness=0.9,
    contrast=0.9,
)


DIR = Path().resolve()
PROJECT_ROOT = Path().resolve().parent.parent

LABEL_CSV = PROJECT_ROOT / "data" / "MGS_data" / "labels" / "labels.csv"
IMAGE_DIR = PROJECT_ROOT / "data" / "MGS_data" / "data"
OUTPATH = DIR / "outputs_color_jittered_grayscale"

def apply_colorjitter(image_path):
    image = Image.open(image_path).convert("RGB")
    image = transform(image)  # Apply color jitter transformation
    return image

df = pd.read_csv(LABEL_CSV)
for file, label in zip(df["filename"], df["label"]):
    image_path = IMAGE_DIR / file
    if not image_path.exists():
        print(f"Image {image_path} does not exist, skipping.")
        continue

    print(f"Processing {image_path} with label {label}")
    transformed_image = apply_colorjitter(image_path)
    
    out_path = OUTPATH / file
    out_path.parent.mkdir(parents=True, exist_ok=True)
    transformed_image.save(out_path)
    print(f"Saved processed image to {out_path}")
    
```
